## **Count Nulls**

```python
df.isnull().sum()
```

## **Remove leading and trailing spaces from all column names**

```python
df.columns = df.columns.str.strip()
```

# **Remove leading and trailing spaces from values in all columns**

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

# **Get a list of column names with null values**

```python
# List of column names
column_names = [
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Changes in inventories',
    'Exports of goods and services',
    'General government final consumption expenditure',
    'Gross capital formation',
    'Gross fixed capital formation (including Acquisitions less disposals of valuables)',
    'Household consumption expenditure (including Non-profit institutions serving households)',
    'Imports of goods and services',
    'Manufacturing (ISIC D)',
    'Transport, storage and communication (ISIC I)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)'
]
```

## **Fill Missing Values**
There are total `11` columns containing missing values. So lets replace them with mean.

```python
for column_name in column_names:
    df[column_name].fillna(df[column_name].mean(), inplace=True)
```

**Let's create some visualization to analyse Pakistan's data.**

## **Per Capita Gross National Income Year Wise**

```python
# Filter the data for Pakistan
pakistan_data = df[df['Country'] == 'Pakistan']
```

```python
# Create line plot
fig = px.line(pakistan_data, x='Year', y='Per capita GNI', title='Per Capita GNI in Pakistan Over the Years')

fig.show()
```

**The graph is showing that **Per Capita GNI** of Pakistan is increasing over a period of Time.**

## **Per Capita GNI Comparison in Asian Countries and Pakistan**

```python
# List of Asian countries
asian_countries = [
    "China", "India", "Japan", "South Korea", "Thailand",
    "Indonesia", "Vietnam", "Malaysia", "Philippines", "Singapore",
    "Pakistan", "Bangladesh", "Sri Lanka", "Nepal", "Afghanistan",
    "Saudi Arabia", "Iran", "Turkey", "United Arab Emirates", "Israel"
]

# the data to include only Asian countries and Pakistan
selected_data = df[(df['Country'].isin(asian_countries)) | (df['Country'] == 'Pakistan')]

# line plot
fig = px.line(selected_data, x='Year', y='Per capita GNI', color='Country', title='Per Capita GNI Comparison in Asian Countries and Pakistan')

fig.show()
```

**The Per Capita GNI of _Pakistan_ is far less than other Asian Countries**

## **Population Trend Line**

```python
# Create line plot
fig = px.line(pakistan_data, x='Year', y='Population', title='Population of Pakistan Over the Years')

fig.show()
```

## **Population Comparison in Asian Countries**

```python
# List of Asian countries
asian_countries = [
    "China", "India", "Japan", "South Korea", "Thailand",
    "Indonesia", "Vietnam", "Malaysia", "Philippines", "Singapore",
    "Pakistan", "Bangladesh", "Sri Lanka", "Nepal", "Afghanistan",
    "Saudi Arabia", "Iran", "Turkey", "United Arab Emirates", "Israel"
]

selected_data = df[(df['Country'].isin(asian_countries)) | (df['Country'] == 'Pakistan')]

# line plot
fig = px.line(selected_data, x='Year', y='Population', color='Country', title='Population Comparison in Asian Countries')

fig.show()
```

## **Bar graph to show population**

```python
# Create a bar graph to show population
fig = px.bar(selected_data, x='Country', y='Population', title='Population of Asian Countries and Pakistan')

# Show the bar graph
fig.show()
```

**And here, we can see that Pakistan is the fourth most populous country in Asia.**

## **World Population by Country**

```python
#  map chart
fig = px.choropleth(df, 
                    locations='Country', 
                    locationmode='country names',
                    color='Population',
                    hover_name='Country',
                    title='World Population by Country',
                    width=1000,  # Set the width
                    height=600)  # Set the height

# Customize the the map
fig.update_geos(showcoastlines=True, coastlinecolor="Black")
fig.update_coloraxes(colorscale="Viridis")

fig.show()
```

## **Import & Export Comparison**

```python
# compare exports and imports
fig = px.bar(pakistan_data, x='Year', y=['Exports of goods and services', 'Imports of goods and services'],
              title='Comparison of Exports and Imports of Goods and Services in Pakistan Over the Years')

# Show the plot
fig.show()
```

**This stacked Bar Graph is indicating that in the year of 2020 the imports of Pakistan is more than 80 Billion and imports are only about 31 Billion.**

## **Comparing Import & Export with India**

```python
# Filter data
selected_data = df[df['Country'].isin(['Pakistan', 'India'])]

# Creating line plot 
fig = px.line(selected_data, x='Year', y=['Exports of goods and services', 'Imports of goods and services'],
              color='Country',
              title='Comparison of Exports and Imports of Goods and Services in Pakistan and India Over the Years')

# rabbit icon for India
fig.add_annotation(
    x=selected_data[selected_data['Country'] == 'India']['Year'].max(),
    y=selected_data[selected_data['Country'] == 'India']['Exports of goods and services'].max(),
    text="India 🐇",
    showarrow=True,
    font=dict(size=30),
)

# turtle icon for Pakistan
fig.add_annotation(
    x=selected_data[selected_data['Country'] == 'Pakistan']['Year'].max(),
    y=selected_data[selected_data['Country'] == 'Pakistan']['Imports of goods and services'].max(),
    text="Pakistan 🐢",
    showarrow=True,
    font=dict(size=30),
)

# Show the plot
fig.show()
```

## **GDP Comparison Between Pakistan and India**

```python
# only Pakistan and India
selected_data = df[df['Country'].isin(['Pakistan', 'India'])]

# line plot 
fig = px.line(selected_data, x='Year', y='Gross Domestic Product (GDP)', color='Country',
              title='GDP Comparison Between Pakistan and India Over Time')

fig.show()
```

```python
# Asian countries
selected_data = df[df['Country'].isin(asian_countries)]

# a bar chart to show GDP
fig = px.line(selected_data, x='Year', y='Gross Domestic Product (GDP)', color='Country',
             title='GDP Comparison in Asian Countries Over Time')

fig.show()
```

## **Correlation Heat Map**

```python
# Select the relevant columns
selected_data = pakistan_data[['Year', 'Imports of goods and services', 'Exports of goods and services', 'Gross Domestic Product (GDP)']]

# Calculate the correlation matrix
correlation_matrix = selected_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap for Import, Export, and GDP in Pakistan')
plt.show()
```

## **Stacked Bar Chart for Import, Export, and GDP in Pakistan**

```python
# Select the relevant columns
selected_data = pakistan_data[['Year', 'Imports of goods and services', 'Exports of goods and services', 'Gross Domestic Product (GDP)']]

# Set the 'Year' column as the index (if it's not already)
selected_data.set_index('Year', inplace=True)

# Create a stacked bar chart
ax = selected_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart for Import, Export, and GDP in Pakistan')
plt.xlabel('Year')
plt.ylabel('Values')

# Show the legend
plt.legend(loc='upper right')

plt.show()
```
